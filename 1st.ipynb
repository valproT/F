```python
import numpy as np
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def sigmoid_dervative(x):
    return x*(1-x)
```

```python
X=np.array([[0,0],
            [0,1],
            [1,0],
            [1,1]])
y= np.array([[0],[0],[0],[1]])

#wts and bais
weights=np.random.rand(2,1)
bias=np.random.rand(1)
print("Initial wts:",weights)
print("Initial bias:", bias)
```

```python
learning_rate=0.01
epochs=1000
```

```python
print("Testing AND gate before training:")
for i in range(len(X)):
    input_data=X[i]
    output=sigmoid(np.dot(input_data,weights)+bias)
    print(f"INPUT:{input_data},normal_output:{output},Output:{np.round(output)}")
```

```python
for epoch in range(epochs):
    input_layer=X
    weighted_sum=np.dot(input_layer,weights)+bias
    activated_output=sigmoid(weighted_sum)
    print(activated_output)
    #error
    error=y-activated_output

    #adjust the backpropagation
    adjust=error*sigmoid_dervative(activated_output)
    weights+=np.dot(input_layer.T,adjust)*learning_rate
    bias+=np.sum(adjust)*learning_rate
    
```

```python
print("Final weights:",weights)
print("Final bias:",bias)
```

```python
#testing the AND gate
for i in range(len(X)):
    input_data=X[i]
    output=sigmoid(np.dot(input_data,weights)+bias)
    print(f"Input:{input_data},Output:{np.round(output).astype(int)}")
```
